```python
%%writefile tmin_mean_300.py
"""
compute mean vector of snow for all the stations with 365 values
"""
import sys
sys.path.append('/usr/lib/python2.7/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr
import pickle, gzip
import pandas
import numpy as np

class SnowMean(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            elements=line.split(',')       
            if elements[0]=='station':
                yield ('header',1)
            else:
                if elements[1]=='TMIN':
                    #mean_yr = 0
                    mvalues = []
                    i=0
                    for e in elements[3:]:
                    # Added Threshold:
                        if e!='':
                            mvalues.append(float(e))
                            i = i + 1
                        else:
                            mvalues.append(np.nan)
                    mvalues = np.array(mvalues)
                    if i >= 300:
                        M_mask = np.ma.masked_array(mvalues,np.isnan(mvalues))
                        M_mean = np.mean(M_mask)
                        yield(elements[0],float(M_mean))
                    #isnan = 0
                    #for i in xrange(3,int(len(elements))):
                     #   if elements[i] != '':
                      #      mean_yr += float(elements[i])
                       # else:
                        #    isnan = 1
                         #   break
                            
                    #if isnan == 1:
                     #   yield('other',1)
                    #else:
                     #   yield (elements[0],mean_yr/365)
                else:
                    yield ('other',1)
        except Exception, e:
            stderr.write('Error in line:\n'+line+'\n')
            stderr.write(str(e)+'\n')
            self.increment_counter('MrJob Counters','mapper-error',1)
            yield ('error',1)

    def reducer(self, station, mean_yrs):
        self.increment_counter('MrJob Counters','reducer',1)
        if station!='other':
            mean_yrs = (list(mean_yrs))
            yield (station,sum(mean_yrs)/len(mean_yrs))

if __name__ == '__main__':
    SnowMean.run()
```
